```python
number_ratings = ratings_with_books.groupby('title')['rating'].count().reset_index().rename(columns={'rating': 'numberOfRatings'})
```

```python
number_ratings
```

```python
final_ratings = ratings_with_books.merge(number_ratings, on='title' )
```

```python
final_ratings
```

```python
final_ratings['numberOfRatings'] > 30
```

```python
final_ratings = final_ratings[final_ratings['numberOfRatings'] >=50]
```

```python
final_ratings
```

```python
final_ratings.shape
```

```python
final_ratings.drop_duplicates(['userID', 'title'], inplace=True)
```

```python
final_ratings.shape
```

```python
book_pivot = final_ratings.pivot_table(columns = 'userID', index= 'title', values= 'rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0 , inplace= True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
books_sparse = csr_matrix(book_pivot)
```

```python
books_sparse
```

```python
type(books_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(books_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[125, :].values.reshape(1,-1), n_neighbors = 6)
```

```python
distances
```

```python
suggestions
```
